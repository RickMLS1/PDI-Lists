# **Create Images**

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Create the image of a paraboloid with one axis scaled (like an oval paraboloid).

```python
n_de_pontos = 12
x = np.linspace(-6, 6, n_de_pontos, endpoint=True)
y = np.linspace(-6, 6, n_de_pontos, endpoint=True)
X, Y = np.meshgrid(x, y)

Z = 2*X ** 2 + Y ** 2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='Dark2')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Paraboloide Oval')

plt.show()
```

# Create the image of a rotated sin using rotation of coordinates.

```python
n_de_pontos = 60
x = np.linspace(-6, 6, n_de_pontos, endpoint=True)
y = np.sin(x)

teta = np.pi / 4

x_rotacionado = x * np.cos(teta) + y * np.sin(teta) # x' = x cos(teta) + y sin(teta)

y_rotacionado = -x * np.sin(teta) + y * np.cos(teta) # y' = -x sin(teta) + y cos(teta)

plt.plot(x_rotacionado, y_rotacionado)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Seno rotacionado')
plt.grid(True)
plt.show()
```

# Create the image of a gaussian.

```python
media = 0
desvio_padrao = 1
n_de_pontos = 60

x = np.linspace(-6, 6, n_de_pontos, endpoint=True)
y = np.exp(-((x - media) ** 2) / (2 * desvio_padrao ** 2)) / (desvio_padrao * np.sqrt(2 * np.pi))

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian')
plt.grid(True)

plt.show()
```

# Create a function that generates the image of a Gaussian optionally rotated by an angle theta and with mx, my, sx, sy as input arguments.

```python
mx = float(input("Digite o valor de mx: ")) # translate in eixo - x
my = float(input("Digite o valor de my: ")) # translate in eixo - y
sx = float(input("Digite o valor de sx: ")) # escala no eixo - x
sy = float(input("Digite o valor de sy: ")) # escala no eixo - y

media = 0
desvio_padrao = 1
teta = np.pi / 4 # angulo de rotacao
n_de_pontos = 60

x = np.linspace(-6, 6, n_de_pontos, endpoint=True)
y = np.exp(-((x - media) ** 2) / (2 * desvio_padrao ** 2)) / (desvio_padrao * np.sqrt(2 * np.pi))

x_rotacionado = x * np.cos(teta) + y * np.sin(teta) # x' = x cos(teta) + y sin(teta)
y_rotacionado = -x * np.sin(teta) + y * np.cos(teta) # y' = -x sin(teta) + y cos(teta)

x_transformado = (x_rotacionado * sx) + mx
y_transformado = (y_rotacionado * sy) + my

plt.plot(x_transformado, y_transformado)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian')
plt.grid(True)

plt.show()
```
